### Functions: def and apply

```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
url = "https://data.cityofnewyork.us/api/views/gaq9-z3hz/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)
```

```python
df.head(3)
```

```python
df['Ratio'] = df['Capture Rate-Paper (Total Paper / Max Paper)']/df['Capture Rate-MGP (Total MGP / Max MGP)']
```

```python
df.head(3)
```

```python
def compare(p):
    if p > 1:
        return True
    else:
        return False
```

```python
df['Paper_Wins'] = df['Ratio'].apply(compare)
df
```

```python
def compare(p,a):
    if p > a:
        return True
    else:
        return False
```

```python
df['Paper_Wins_1'] = df.apply(lambda row: compare(row['Ratio'],1), axis=1)
df
```
